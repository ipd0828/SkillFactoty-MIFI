Сколькими способами можно поставить 5 различных книг на полке, если две из них должны стоять рядом?

Рассмотрим две книги, которые должны стоять рядом, как один "блок". Тогда у нас будет 4 "объекта" для размещения на полке: этот "блок" и остальные 3 книги.

Число способов разместить 4 "объекта" на полке равно $4!$, что составляет:

$
4! = 4 \times 3 \times 2 \times 1 = 24
$

Внутри "блока" две книги можно переставить между собой $2!$ способами:

$
2! = 2 \times 1 = 2
$

Общее число способов расставить книги, чтобы две из них стояли рядом, равно:

$
4! \times 2! = 24 \times 2 = 48
$

Ответ: 48

Чтобы найти вероятность того, что оба телефона работают, нам нужно рассмотреть вероятности, что каждый телефон не сломан.

Вероятность того, что первый телефон работает = $1 - 0.1 = 0.9$

# Задание 13
Сколькими способами можно распределить 10 одинаковых яблок между 4 детьми, если каждый ребенок должен получить хотя бы одно яблоко?

Решим задачу о распределении 10 одинаковых яблок между 4 детьми, при условии, что каждый ребенок должен получить хотя бы одно яблоко. Для этого используем метод "сначала отдать каждому ребенку по одному яблоку".

Шаг 1: Первоначальное распределение

Первый шаг — отдать каждому ребенку по одному яблоку. Поскольку у нас 4 ребенка, мы отдадим 4 яблока. После этого у нас останется:

$$
10 - 4 = 6 \text{ яблок}.
$$

Шаг 2: Оставшиеся яблоки

Теперь нам нужно распределить оставшиеся 6 яблок между 4 детьми. Здесь мы можем использовать метод "разделения яблок", где мы рассматриваем распределение как задачу о неотрицательных целых решениях.

Шаг 3: Формулировка задачи

Нам нужно найти неотрицательные целые числа \(x_1, x_2, x_3, x_4\), которые удовлетворяют уравнению:

$$
x_1 + x_2 + x_3 + x_4 = 6,
$$

где \(x_i\) — это количество дополнительных яблок, которые получает i-й ребенок.

Шаг 4: Использование формулы

Для решения этой задачи мы можем использовать формулу для вычисления количества неотрицательных целых решений уравнения:

$$
\binom{n+k-1}{k-1},
$$

где \(n\) — это количество яблок (в нашем случае 6), а \(k\) — количество детей (в нашем случае 4).

Шаг 5: Подстановка значений

Подставим значения в формулу:

$$
n = 6, \quad k = 4 \quad \Rightarrow \quad \binom{6 + 4 - 1}{4 - 1} = \binom{9}{3}.
$$

 Шаг 6: Вычисление биномиального коэффициента

Вычислим биномиальный коэффициент:

$$
\binom{9}{3} = \frac{9!}{3!(9 - 3)!} = \frac{9 \times 8 \times 7}{3 \times 2 \times 1} = \frac{504}{6} = 84.
$$

Итог

Таким образом, количество способов распределить 10 одинаковых яблок между 4 детьми, если каждый ребенок должен получить хотя бы одно яблоко, равно

$$
Ответ: 84.
$$

# Задание 15

Монету бросают 6 раз. Какова вероятность того, что орел выпадет ровно 4 раза?

Чтобы найти вероятность того, что орел выпадет ровно 4 раза при 6 бросках монеты, используем биномиальное распределение. Вероятность успеха (выпадения орла) в каждом броске равна $ p = 0.5 $, так как монета честная. 

Формула биномиального распределения:

$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$

где:
- $ n = 6 $ — общее количество бросков,
- $ k = 4 $ — количество успехов (выпадений орла),
- $ p = 0.5 $ — вероятность выпадения орла.

Подставляем значения в формулу:

$
P(X = 4) = \binom{6}{4} (0.5)^4 (0.5)^{6-4}
$

Вычислим биномиальный коэффициент $\binom{6}{4}$:

$
\binom{6}{4} = \frac{6!}{4!(6-4)!} = \frac{6 \times 5}{2 \times 1} = 15
$

Теперь подставим все значения:

$
P(X = 4) = 15 \times (0.5)^4 \times (0.5)^2 = 15 \times (0.5)^6
$

$
(0.5)^6 = \frac{1}{64}
$

Таким образом:

$
P(X = 4) = 15 \times \frac{1}{64} = \frac{15}{64}
$

Ответ: $\frac{15}{64}$

# Задание 17
В треугольнике ABC известно, что AC = BC, AH - высота, tg(BAC) = 7/24. Найдите косинус угла BAH.

Запшем условия задачи глядя на построение чуть иначе и более подробно.Рассмотрим треугольник ABC, в котором AC = BC. Обозначим AH — высоту, опущенную из точки A на сторону BC. В прямоугольном треугольнике AHB угол AHB является прямым. На сторону HB опущен луч AC, что образует равнобедренный треугольник ACB, где AC = BC. Дано, что

$$
\tan(BAC) = \frac{7}{24}.
$$

Необходимо найти косинус угла BAH.

<b>Аналитическое решение</b>

1. Обозначим углы и стороны:
   - Обозначим угол BAC как \(\alpha\).
   - В соответствии с заданным значением, имеем

$$
\tan(\alpha) = \frac{7}{24}.
$$

2. Выражение высоты AH:
   - В прямоугольном треугольнике AHB, используя определение тангенса, можем записать:

$$
\tan(\alpha) = \frac{AH}{BH} \implies AH = \tan(\alpha) \cdot BH = \frac{7}{24} \cdot BH.
$$

   - Обозначим BH как \(24x\). Тогда:

$$
AH = 7x.
$$

3. Находим сторону AB с использованием теоремы Пифагора:
   - В треугольнике AHB применяем теорему Пифагора:

$$
AB^2 = AH^2 + BH^2 = (7x)^2 + (24x)^2 = 49x^2 + 576x^2 = 625x^2.
$$

   - Следовательно,

$$
AB = 25x.
$$

4. Теперь можем найти косинус угла BAH:
   - Используя определение косинуса, получаем:

$$
\cos(BAH) = \frac{AH}{AB} = \frac{7x}{25x} = \frac{7}{25}.
$$

5. Вычисление:
   - Приведем значение к десятичной форме:

$$
\cos(BAH) = \frac{7}{25} = 0.28.
$$

<b>Ответ</b>

Косинус угла BAH равен \(0.28\).

# Задание 19
Найдите объём правильной треугольной пирамиды, стороны основания которой равны 1, а высота равна $\sqrt{3}$

Используем формулу:

$
V = \frac{1}{3} S h
$

где $ S $ — площадь основания, а $ h $ — высота пирамиды.

Основание пирамиды является правильным треугольником со стороной $ a = 1 $.

Площадь правильного треугольника вычисляется по формуле:

$
S = \frac{\sqrt{3}}{4} a^2
$

Подставим значение стороны:

$
S = \frac{\sqrt{3}}{4} \cdot 1^2 = \frac{\sqrt{3}}{4}
$


Высота $ h = \sqrt{3} $:

$
V = \frac{1}{3} S h = \frac{1}{3} \cdot \frac{\sqrt{3}}{4} \cdot \sqrt{3}
$

$
V = \frac{1}{3} \cdot \frac{\sqrt{3} \cdot \sqrt{3}}{4} = \frac{1}{3} \cdot \frac{3}{4} = \frac{3}{12} = \frac{1}{4}
$


Ответ: $\frac{1}{4}$

В равнобедренной трапеции основания равны 12 и 27, острый угол равен 60 градусам.

Найдите её периметр

Назовем трапецию ABCD, где A - острый угол = D

Из вершины В опустим высоту ВЕ

$АЕ = (AD - BC) / 2$

$AE = (27 - 12) / 2 = 15 / 2 = 7.5$

Рассмотрим прямоугольный треугольник АВЕ:

Угол А = 60 градусов, значит угол В = 90-60 = 30 градусов

Катет, лежащий против угла в 30 градусов равен половине гипотенузы, значит $АВ = 2 * АЕ$

$АВ = 2 * 7.5 = 15$

Периметр трапеции равен:
