Исследуем данные на наличие линейной зависимости между столбцами. Найдите ранг матрицы данных. Сравните его с количеством столбцов, сделайте вывод. Если линейная зависимость есть - напишите, между какими столбцами и удалите один из них.

```python
import numpy as np
```

```python
# рассчитаем корреляцию между числовыми столбцами
correlation_matrix = data.corr()
print('Корреляционная матрица: ')
correlation_matrix
```

```python
# найдем ранг матрицы
matrix_rank = np.linalg.matrix_rank(data.select_dtypes(include=[np.number]))
print("Ранг матрицы:", matrix_rank)
```

```python
# cравним ранг с количеством столбцов
num_columns = data.select_dtypes(include=[np.number]).shape[1]
print("Количество столбцов:", num_columns)
```

```python
# вывод по линейной зависимости
if matrix_rank < num_columns:

    print("Есть линейная зависимость между некоторыми столбцами")

    # Находим пары столбцов с высокой корреляцией (выше 0.8)
    high_correlation = correlation_matrix[
        (correlation_matrix > 0.8) & (correlation_matrix < 1.0)
    ]
    print("\nСтолбцы с высокой корреляцией:")
    display(high_correlation)

else:
    print("\nНет линейной зависимости между столбцами")
```

На основе полученной корреляционной матрицы видно, что значения корреляции между некоторыми столбцами очень высоки

- temp и atemp: корреляция 0.991702 — зависимость между температурой и температурой по ощущениям.
- windspeed(mph) и windspeed(ms): корреляция 1.0 — полная зависимость, так как скорость ветра измеряется в двух единицах.

## Задание 10
Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
X = data.drop(columns=["cnt"])  # все переменные, кроме целевой переменной
y = data["cnt"]  # целевая переменная
```

```python
# столбец единичных значений для смещения (bias)
X = np.c_[np.ones(X.shape[0]), X]

X = np.array(X)
y = np.array(y)

# находим коэффициенты линейной регрессии
beta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
# получаем прогнозы
y_pred = X @ beta
```

```python
# Среднеквадратичная ошибка
mse = np.mean((y - y_pred) ** 2)

# Коэффициент детерминации R^2
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_residual / ss_total)

print(f"MSE: {mse}")
print(f"R^2: {r2}")
```
