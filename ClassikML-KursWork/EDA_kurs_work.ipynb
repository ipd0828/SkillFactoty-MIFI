## Параметры оценки противовирусной активности: IC50, CC50, SI

### Ключевые показатели эффективности
В исследованиях противовирусных соединений, особенно против вируса гриппа, используются три фундаментальных параметра:

1. **IC50 (Half Maximal Inhibitory Concentration)**  
   - Концентрация, необходимая для подавления вирусной репликации на 50%  
   - **Интерпретация**: Чем **ниже** значение, тем выше противовирусная активность  
   - Единицы: μM или nM  
   - *Пример*: Соединение с IC50=0.1 μM эффективнее, чем с IC50=10 μM

2. **CC50 (Half Maximal Cytotoxic Concentration)**  
   - Концентрация, вызывающая гибель 50% здоровых клеток  
   - **Интерпретация**: Чем **выше** значение, тем ниже токсичность  
   - Единицы: μM или nM (согласованы с IC50)  
   - *Пример*: Соединение с CC50=100 μM безопаснее, чем с CC50=5 μM

3. **SI (Selectivity Index)**  
   - Расчетный параметр: **SI = CC50 / IC50**  
   - **Интерпретация**:  
     - Показатель терапевтического окна  
     - Чем **выше** значение, тем лучше профиль безопасности  
     - Минимальный порог для перспективных соединений: SI > 8  
   - *Пример*: При CC50=100 μM и IC50=1 μM → SI=100 (отличный показатель)

```python
# Функции
def check_corr_method(data, method='pearson', target='target', count=10):
    """
    Функция для расчета корреляции 'kendall','pearson','spearman' всех признаков с целевой переменной.
    ---
    Принимает на вход pd.DataFrame.
    Возвращает pd.Series с значением коэффициента корреляции для каждого признака.
    """
    # чекаем корреляцию с таргетом
    correlation = data.iloc[:, :-1].corrwith(data[target], method=method)
    
    # сортируем
    correlation = abs(correlation.sort_values(ascending=False))
    
    return correlation[:count]
```

```python
# для проверки берем все признаки за исключением идентификатора субъекта
corr_features = df.copy()
# строим матрицу корреляций
corr_matrix = corr_features.corr()

# порог отсечения корреляции
threshold = 0.95
# здесь соберем признаки для удаления

col1_list = []
col2_list = []
# итерируемся по столбцам
for i, col1 in enumerate(corr_features.columns):
    for j, col2 in enumerate(corr_features.columns):
        # если коэффициент корреляции для этой пары больше порога
        if i < j and abs(corr_matrix.iloc[i, j]) > threshold:
            # выводим в консоль пару
            print(f'pair: 1={col1} and 2={col2} corr={abs(corr_matrix.iloc[i, j])}')
            # добавляем столбец в коллекцию на удаление
            
            col1_list.append(col1)
            col2_list.append(col2)

# выводим в консоль признаки, отобранные для удаления
to_drop = list(set(col1_list).difference(set(col2_list)))

print(to_drop)
```

```python
# Собрал различные вариации новых признаков , смысл котрых нашли в интернете знающие люди
def add_engineered_features(df):
    
    df = df.copy()
    
    # 1. Комбинированные физико-химические свойства
    df['LogP_MW_Ratio'] = df['MolLogP'] / (df['MolWt'] + 1e-6)
    df['PolarSurfaceFraction'] = df['TPSA'] / (df['LabuteASA'] + 1e-6)
    df['Charge_Asymmetry'] = df['MaxPartialCharge'] - df['MinPartialCharge']
    
    # 2. Топологические индексы
    df['Complexity_Index'] = df['BertzCT'] * df['HallKierAlpha'] / (df['NumRotatableBonds'] + 1)
      
    # 3. Электронные свойства
    df['BCUT_Charge_Ratio'] = df['BCUT2D_CHGHI'] / (abs(df['BCUT2D_CHGLO']) + 1e-6)
    df['EState_Range'] = df['MaxEStateIndex'] - df['MinEStateIndex']
    
    # 4. Функциональные группы
    polar_groups = ['fr_COO', 'fr_Ar_OH', 'fr_Al_OH', 'fr_NH2', 'fr_amide']
    lipo_groups = ['fr_halogen', 'fr_alkyl_halide', 'fr_unbrch_alkane', 'fr_aryl_methyl']
    
    df['Polar_Groups_Count'] = df[polar_groups].sum(axis=1)
    df['Lipophilic_Groups_Count'] = df[lipo_groups].sum(axis=1)
    df['PLB'] = df['Polar_Groups_Count'] / (df['Lipophilic_Groups_Count'] + 1e-6)
    
    # 5. Водородные связи
    df['HBond_Capacity'] = df['NumHDonors'] + df['NumHAcceptors']
    df['HBond_Donor_Acceptor_Ratio'] = df['NumHDonors'] / (df['NumHAcceptors'] + 1e-6)
    
    # 6. Циклические системы
    df['Saturation_Index'] = df['NumSaturatedRings'] / (df['RingCount'] + 1e-6)
    
    # 7. Производные от VSA
    hydrophobic_vsa = [f'SlogP_VSA{i}' for i in range(1, 7)]
    if set(hydrophobic_vsa).issubset(df.columns):
        df['Hydrophobic_VSA'] = df[hydrophobic_vsa].sum(axis=1)
    
    # 8. BCUT взаимодействия
    df['BCUT_Electronic'] = df['BCUT2D_MWHI'] * df['BCUT2D_CHGHI']

    # 9. Дополнительные найденные варианты
    df['Size_Flexibility'] = df['MolWt'] * df['NumRotatableBonds'] / 100
    df['LogD'] = df['MolLogP'] - np.log10(df['Polar_Groups_Count'] + 1)
    
    return df
```

```python
# Перепроверяем
bool_features = [col for col in bool_features if col in data.columns]
cat_features = [col for col in cat_features if col in data.columns]

# Числовые признаки (исключая целевые, булевы и категориальные)
numeric_features = [col for col in data.select_dtypes(include=[np.number]).columns 
                   if col not in bool_features + cat_features + ['IC50, mM', 'CC50, mM', 'SI']]
```

```python
# Создание препроцессора 
preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numeric_features),
        ('bool', 'passthrough', bool_features),
       ('cat', 'passthrough', cat_features)
    ])

# Функция для подготовки данных
def prepare_datasets(df, target_col, drop_cols):
    # Копируем и удаляем связанные столбцы, предотвращая утечку целевого признака
    df = df.drop(drop_cols, axis=1).copy()
    
    # Создаем целевые переменные
    if target_col == 'IC50, mM':
        y = np.log1p(df[target_col])
        y_clf = (df[target_col] > df[target_col].median()).astype(int)
        return {
            'reg': (df.drop(target_col, axis=1), y),
            'clf': (df.drop(target_col, axis=1), y_clf)
        }
    elif target_col == 'CC50, mM':
        y = np.log1p(df[target_col])
        y_clf = (df[target_col] > df[target_col].median()).astype(int)
        return {
            'reg': (df.drop(target_col, axis=1), y),
            'clf': (df.drop(target_col, axis=1), y_clf)
        }
    else:  # SI
        y = np.log1p(df[target_col])
        y_clf_med = (df[target_col] > df[target_col].median()).astype(int)
        y_clf_8 = (df[target_col] > 8).astype(int)
        return {
            'reg': (df.drop(target_col, axis=1), y),
            'clf_med': (df.drop(target_col, axis=1), y_clf_med),
            'clf_8': (df.drop(target_col, axis=1), y_clf_8)
        }

# Подготовка данных для каждой задачи
ic50_data = prepare_datasets(data, 'IC50, mM', ['CC50, mM', 'SI'])
cc50_data = prepare_datasets(data, 'CC50, mM', ['IC50, mM', 'SI'])
si_data = prepare_datasets(data, 'SI', ['IC50, mM', 'CC50, mM'])
```

```python
# Разделение данных на train/test
def split_data(data_dict):
    results = {}
    for key, (X, y) in data_dict.items():
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.15, random_state=Randome_state
        )
        results[key] = (X_train, X_test, y_train, y_test)
    return results

ic50_splits = split_data(ic50_data)
cc50_splits = split_data(cc50_data)
si_splits = split_data(si_data)
```

```python
# модели
models_reg = {
    'KNN': KNeighborsRegressor(),
    'GBR': GradientBoostingRegressor(random_state=Randome_state),
    'XGB': XGBRegressor(random_state=Randome_state),
    'RF': RandomForestRegressor(random_state=Randome_state, n_jobs=14),
    'SVR': SVR(),
    'CAT': CatBoostRegressor(random_state=Randome_state, verbose=False)
}

models_clf = {
    'KNN': KNeighborsClassifier(),
    'GBC': GradientBoostingClassifier(random_state=Randome_state),
    'XGBC': XGBClassifier(random_state=Randome_state, use_label_encoder=False, eval_metric='logloss'),
    'RFC': RandomForestClassifier(random_state=Randome_state, n_jobs=14),
    'SVC': SVC(probability=True, random_state=Randome_state),
    'CATC': CatBoostClassifier(random_state=Randome_state, auto_class_weights='Balanced', verbose=False)
}
```

```python
# сетки параметров
param_grids = {
    'KNN': {'model__n_neighbors': [3, 5, 7], 'model__weights': ['uniform', 'distance']},
    'GBR': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1], 'model__max_depth': [3,  12]},
    'XGB': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1], 'model__max_depth': [3,  12]},
    'RF': {'model__n_estimators': [100, 200], 'model__max_depth': [5, 10]},
    'SVR': {'model__C': [0.1, 1], 'model__epsilon': [0.1, 0.5]},
    'CAT': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1]},
    'GBC': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1], 'model__max_depth': [3, 12]},
    'XGBC': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1], 'model__max_depth': [3, 12]},
    'RFC': {'model__n_estimators': [100, 200], 'model__max_depth':  [3,  12]},
    'SVC': {'model__C': [0.1, 1], 'model__gamma': ['scale', 'auto']},
    'CATC': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1], 'model__max_depth': [3,12]}
}
```

```python
# Функция для обучения и оценки
def train_and_evaluate(models, param_grids, preprocessor, data_splits, task_type='reg'):
    results = {}
    for name in models:
        # пайплайн
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('model', models[name])
        ])
        
        # (GridSearch для KNN/SVM, RandomizedSearch для остальных)
        if name in ['KNN', 'SVR', 'SVC']:
            search = GridSearchCV(pipeline, param_grids[name], cv=3, n_jobs=14,
                                 scoring='neg_mean_squared_error' if task_type == 'reg' else 'accuracy')
        else:
            search = RandomizedSearchCV(pipeline, param_grids[name], n_iter=5, cv=3,
                                       n_jobs=14, random_state=Randome_state,
                                       scoring='neg_mean_squared_error' if task_type == 'reg' else 'accuracy')
        
        # Обучение и оценка 
        task_results = {}
        for task_name, (X_train, X_test, y_train, y_test) in data_splits.items():
            search.fit(X_train, y_train)
            best_model = search.best_estimator_
            y_pred = best_model.predict(X_test)
            
            if task_type == 'reg':
                mse = mean_squared_error(y_test, y_pred)
                r2 = r2_score(y_test, y_pred)
                task_results[task_name] = {'MSE': mse, 'R2': r2}
            else:
                acc = accuracy_score(y_test, y_pred)
                f1 = f1_score(y_test, y_pred)
                auc = roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1]) if  (best_model, 'predict_proba') else 0.5
                task_results[task_name] = {'Accuracy': acc, 'F1': f1, 'AUC': auc}
        
        results[name] = task_results
    return results

# Обучение регрессионных моделей
print("Обучение регрессионных моделей...")
reg_results = {
    'IC50': train_and_evaluate(models_reg, param_grids, preprocessor, 
                              {'reg': ic50_splits['reg']}, 'reg'),
    'CC50': train_and_evaluate(models_reg, param_grids, preprocessor, 
                              {'reg': cc50_splits['reg']}, 'reg'),
    'SI': train_and_evaluate(models_reg, param_grids, preprocessor, 
                            {'reg': si_splits['reg']}, 'reg')
}

# Обучение классификационных моделей
print("\nОбучение классификационных моделей...")
clf_results = {
    'IC50': train_and_evaluate(models_clf, param_grids, preprocessor, 
                              {'clf': ic50_splits['clf']}, 'clf'),
    'CC50': train_and_evaluate(models_clf, param_grids, preprocessor, 
                              {'clf': cc50_splits['clf']}, 'clf'),
    'SI_med': train_and_evaluate(models_clf, param_grids, preprocessor, 
                                {'clf_med': si_splits['clf_med']}, 'clf'),
    'SI_8': train_and_evaluate(models_clf, param_grids, preprocessor, 
                              {'clf_8': si_splits['clf_8']}, 'clf')
}

```

```python
#  Посмотрим на результаты моделей
def print_and_plot_best_results(results_dict, metric_name, metric_label):
    
    tasks = []
    best_models = []
    best_scores = []
    all_data = []
    
    print(f"\nЛУЧШИЕ МОДЕЛИ ({metric_label}):")
    for task, models in results_dict.items():
        best_model = None
        best_score = -np.inf if metric_name != 'MSE' else np.inf
        task_data = []
        
        for model_name, scores in models.items():
            for subtask, metrics in scores.items():
                if metric_name in metrics:
                    score = metrics[metric_name]
                    task_data.append((model_name, score))
                    
                    if (metric_name != 'MSE' and score > best_score) or \
                       (metric_name == 'MSE' and score < best_score):
                        best_score = score
                        best_model = model_name
        
        if best_model:
            tasks.append(task)
            best_models.append(best_model)
            best_scores.append(best_score)
            print(f"{task}: {best_model} ({best_score:.2f})")
            
            for model, score in task_data:
                all_data.append({
                    'Task': task,
                    'Model': model,
                    'Metric': metric_label,
                    'Score': score
                })
    
    # Визуализация результатов
    if all_data:
        plt.figure(figsize=(14, 8))
        
        plt.subplot(1, 2, 1)
        df_all = pd.DataFrame(all_data)
        ax = sns.barplot(x='Task', y='Score', hue='Model', data=df_all, )
        plt.title(f'Сравнение моделей ({metric_label})')
        plt.xticks(rotation=45, ha='right')
        plt.ylabel(metric_label)
        
        for p in ax.patches:
            ax.annotate(f"{p.get_height():.2f}", 
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', 
                        xytext=(0, 10), 
                        textcoords='offset points')
        
        # Лучшие модели
        plt.subplot(1, 2, 2)
        df_best = pd.DataFrame({
            'Task': tasks,
            'Model': best_models,
            'Score': best_scores
        })
        ax = sns.barplot(x='Task', y='Score', hue='Model', data=df_best, dodge=False)
        plt.title(f'Лучшие модели по задачам ({metric_label})')
        plt.xticks(rotation=45, ha='right')
        plt.ylabel(metric_label)
        
        # Добавляем значения на столбцы
        for p in ax.patches:
            ax.annotate(f"{p.get_height():.2f}", 
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', 
                        xytext=(0, 10), 
                        textcoords='offset points')
        
        plt.tight_layout()
        plt.show()



# Анализ регрессионных моделей
print_and_plot_best_results(reg_results, 'R2', 'R²')
print_and_plot_best_results(reg_results, 'MSE', 'MSE')

# Анализ классификационных моделей
print_and_plot_best_results(clf_results, 'F1', 'F1-score')
print_and_plot_best_results(clf_results, 'AUC', 'AUC')
```
